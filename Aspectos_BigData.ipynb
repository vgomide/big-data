#**Microsoft Azure Data Management Framework**

  É o framework da Microsoft específico para o ciclo de vida dos dados. Este framework é projetado para ajudar organizações a gerenciar seus dados desde a coleta até a eliminação, garantindo conformidade, segurança e maximização de valor. Aqui estão os principais componentes do framework:


### 1. **Ciclo de Vida dos Dados**
   - **Criação e Coleta de Dados**: Foca na identificação e na captura de dados relevantes de várias fontes.
   - **Armazenamento e Governança**: Inclui o gerenciamento e a organização de dados, assegurando a governança e a conformidade.
   - **Preparação e Transformação**: Envolve o processamento de dados para garantir que sejam limpos, consistentes e prontos para análise.
   - **Análise e Insights**: Foca na análise dos dados para extrair insights significativos e na apresentação dos mesmos através de visualizações e relatórios.
   - **Descarte de Dados**: Inclui políticas para a retenção e eliminação segura de dados.

### 2. **Pilares do Framework**
   - **Confiabilidade**: Garantir a precisão e integridade dos dados.
   - **Segurança**: Proteger os dados contra acessos não autorizados e garantir a privacidade.
   - **Escalabilidade**: Permitir que a infraestrutura de dados cresça conforme as necessidades da organização.
   - **Flexibilidade**: Adaptar-se às mudanças nas necessidades de negócios e tecnologia.


### 3. **Ferramentas e Serviços**
   - **Azure Data Lake Storage**: Para armazenamento de dados em larga escala.
   - **Azure Synapse Analytics**: Para análises integradas que permitem consultas em dados de múltiplas fontes.
   - **Power BI**: Para visualização e compartilhamento de insights de dados.
   - **Azure Machine Learning**: Para aplicar inteligência artificial e aprendizado de máquina nos dados.


### 4. **Governança de Dados**
   - **Azure Purview**: Uma solução de governança de dados que permite catalogar, classificar e gerenciar os dados da organização de forma eficiente.
   - **Políticas de Segurança**: Para garantir que o acesso e a manipulação de dados sigam as diretrizes da empresa e os regulamentos de conformidade.


### **5. Métodos de Coleta de Dados**
### ETL (Extract, Transform, Load)

1. **Processo**:
   - **Extração**: Os dados são extraídos de fontes diversas, como bancos de dados, sistemas de arquivos ou APIs.
   - **Transformação**: Os dados são transformados em um formato adequado para análise durante esta etapa. Isso pode incluir limpeza, agregação, e enriquecimento dos dados.
   - **Carga**: Os dados transformados são então carregados em um data warehouse ou sistema de armazenamento.

2. **Quando a Transformação Acontece**:
   - A transformação ocorre antes da carga dos dados. Os dados são preparados e estruturados antes de serem inseridos no sistema de destino.

3. **Uso Típico**:
   - Utilizado em sistemas de data warehousing tradicionais, onde a estrutura e a qualidade dos dados são críticas antes da carga.
   - Comumente empregado em ambientes de processamento em lote, onde a transformação dos dados pode levar tempo.

4. **Ferramentas**:
   - Ferramentas populares de ETL incluem Talend, Apache Nifi, Informatica, e Microsoft SSIS.

### ELT (Extract, Load, Transform)

1. **Processo**:
   - **Extração**: Semelhante ao ETL, os dados são extraídos de fontes.
   - **Carga**: Os dados são carregados diretamente no sistema de armazenamento (como um data lake ou um data warehouse).
   - **Transformação**: A transformação ocorre após a carga, utilizando o poder de processamento do sistema de destino.

2. **Quando a Transformação Acontece**:
   - A transformação ocorre **depois** da carga dos dados. Os dados brutos são carregados primeiro e podem ser transformados conforme necessário.

3. **Uso Típico**:
   - Frequentemente utilizado em ambientes de big data e data lakes, onde a flexibilidade e a velocidade de processamento são necessárias.
   - Permite que as empresas armazenem dados em seu formato original e realizem transformações conforme as necessidades de análise mudam.

4. **Ferramentas**:
   - Ferramentas populares de ELT incluem Azure Data Lake, Google BigQuery, Amazon Redshift e Snowflake.

### Resumo

- **ETL** é ideal para ambientes onde a qualidade e a estrutura dos dados precisam ser garantidas antes da carga, enquanto **ELT** é mais adequado para cenários que exigem maior flexibilidade e capacidade de trabalhar com grandes volumes de dados de forma rápida.
- **ETL** transforma dados antes da carga, enquanto **ELT** carrega dados brutos primeiro e realiza transformações posteriormente.

# **Tipos de Gráficos**

## **1. Gráficos de Barras**

*   Uso: Comparar valores entre diferentes categorias.
*   Vantagens: Fácil de entender, ideal para mostrar comparações.




```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulando dados
categorias = ['A', 'B', 'C', 'D', 'E', 'serviço F', 'G']
valores = np.random.randint(10, 100, size=len(categorias))
#valores = np.random.randint(10, 100, size=6)
df_barras = pd.DataFrame({'Categoria': categorias, 'Valores': valores})

# Criando gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Categoria', y='Valores', data=df_barras, hue='Categoria', palette='viridis', legend=False)
plt.title('Gráfico de Barras - Valores por Categoria', fontsize=16)
plt.xlabel('Categoria', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.show()
```

## **2. Gráficos de Linhas**
*   Uso: Mostrar tendências ao longo do tempo.
*   Vantagens: Útil para visualizar séries temporais e identificar padrões




```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Simulando dados para o gráfico de linhas
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Criando o DataFrame
data = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})

# Criando o gráfico de linhas com Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='x', y='y1', data=data, label='Seno')
sns.lineplot(x='x', y='y2', data=data, label='Cosseno')
plt.title('Gráfico de Linhas com Seaborn')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

## **3. Gráficos de Dispersão (Scatter Plots)**
*   Uso: Mostrar a relação entre duas variáveis.
*   Vantagens: Excelente para identificar correlações e padrões em grandes conjuntos de dados.


```python
# Simulando dados
x = np.random.rand(100)
y = np.random.rand(100)

# Criando gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, hue=x, palette='Reds', s=100, alpha=0.7, legend=False)
plt.title('Gráfico de Dispersão - Relação entre X e Y', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.grid()
plt.show()
```

## **4. Histogramas**
*   Uso: Mostrar a distribuição de uma variável.
*   Vantagens: Útil para entender a frequência de dados dentro de intervalos.



```python
# Simulando dados
dados = np.random.randn(1000)

# Criando histograma
plt.figure(figsize=(10, 6))
sns.histplot(dados, bins=30, color='purple', kde=True, legend=False)
plt.title('Histograma - Distribuição de Dados', fontsize=16)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.grid()
plt.show()
```

## **5. Box Plots (Gráficos de Caixa)**
*   Uso: Resumir a distribuição de um conjunto de dados.
*   Vantagens: Útil para identificar outliers e comparar distribuições entre diferentes grupos.



```python
# Simulando dados
dados_box = pd.DataFrame({
    'Grupo': np.random.choice(['A', 'B', 'C'], size=100),
    'Valores': np.random.randn(100)
})

# Criando box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Grupo', y='Valores', data=dados_box, hue='Grupo', palette='Set2', legend=False)
plt.title('Box Plot - Distribuição de Valores por Grupo', fontsize=16)
plt.xlabel('Grupo', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.grid()
plt.show()
```

## **6. Gráficos de Área**
*   Uso: Mostrar a evolução de dados ao longo do tempo, similar aos gráficos de linha.
*   Vantagens: Útil para enfatizar a magnitude das mudanças ao longo do tempo.



```python
# Simulando dados
tempo_area = np.arange(1, 11)
valores_area = np.random.randint(10, 100, size=len(tempo_area))

# Criando gráfico de área
plt.figure(figsize=(10, 6))
plt.fill_between(tempo_area, valores_area, color='skyblue', alpha=0.4)
plt.plot(tempo_area, valores_area, color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Gráfico de Área - Evolução de Valores', fontsize=16)
plt.xlabel('Tempo', fontsize=12)
plt.ylabel('Valores', fontsize=12)
plt.grid()
plt.show()
```

## **7. Heatmaps**
*   Uso: Representar dados em duas dimensões, onde os valores são representados por cores.
*   Vantagens: Excelente para visualizar padrões de densidade e relações complexas entre variáveis.



```python
import numpy as np

np.random.seed(43)
# Simulando dados
dados_heatmap = np.random.rand(10, 12)

# Criando heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(dados_heatmap, cmap='YlGnBu', annot=True)
plt.title('Heatmap - Dados Aleatórios', fontsize=16)
plt.xlabel('Colunas', fontsize=12)
plt.ylabel('Linhas', fontsize=12)
plt.show()
```

## **8. Gráficos de Radar (ou Polares)**
*   Uso: Comparar várias variáveis em relação a um ponto central.
*   Vantagens: Útil para representar a performance de múltiplas variáveis em um único gráfico.


```python
from math import pi
import matplotlib.pyplot as plt

# Simulando dados
categorias_radar = ['A', 'B', 'C', 'D', 'E']
valores_radar = [3, 2, 5, 4, 3]

# Criando gráfico de radar
N = len(categorias_radar)
angles = [n / float(N) * 2 * pi for n in range(N)]
valores_radar += valores_radar[:1]
angles += angles[:1]

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
plt.xticks(angles[:-1], categorias_radar)

ax.plot(angles, valores_radar, color='orange', linewidth=2, linestyle='solid')
ax.fill(angles, valores_radar, color='orange', alpha=0.25)
plt.title('Gráfico de Radar - Comparação entre Categorias', fontsize=16)
plt.show()
```

## **9. Gráficos de Rede (Network Graphs)**
*   Uso: Representar relacionamentos e conexões entre entidades.
*   Vantagens: Ideal para visualizar dados de redes sociais, relacionamentos complexos e conexões entre dados.


```python
import networkx as nx

# Criando um gráfico de rede simples
G = nx.erdos_renyi_graph(10, 0.5)

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_color='black', font_weight='bold', edge_color='gray')
plt.title('Gráfico de Rede - Conexões entre Nós', fontsize=16)
plt.show()
```

# **Agrupamento (Clusterização)**

## **1. K-Means Clustering**

```python
import numpy as np
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Simulando dados
np.random.seed(42)
X = np.random.rand(100, 2) * 100

# Aplicando K-Means
kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X)

# Visualizando os resultados
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=100, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroides')
plt.title('K-Means Clustering', fontsize=16)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

## **2. Hierarchical Clustering**

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Aplicando clusterização hierárquica
linked = linkage(X, method='ward')

# Visualizando o dendrograma
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title('Dendrograma - Agrupamento Hierárquico', fontsize=16)
plt.xlabel('Clusters', fontsize=12)
plt.ylabel('Distância', fontsize=12)
plt.grid()
plt.show()
```

## **3. Método do Cotovelo (Elbow Method)**

### Para determinar o número ideal de clusters usando o Método do Cotovelo (Elbow Method), analisamos o gráfico que relaciona a soma dos quadrados das distâncias (inertia) em função do número de clusters (k). O ponto onde a curva começa a se estabilizar (ou "formar um cotovelo") indica o número

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Simulando dados
np.random.seed(42)
X = np.random.rand(100, 2) * 100

# Calculando a soma dos quadrados da distância (inertia) para diferentes valores de k
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Visualizando o gráfico do Método do Cotovelo
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Método do Cotovelo', fontsize=16)
plt.xlabel('Número de Clusters (k)', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.xticks(k_values)
plt.grid()

# Identificando o número ideal de clusters
ideal_k = 3  # Identificado visualmente onde o "cotovelo" ocorre
plt.axvline(x=ideal_k, color='red', linestyle='--', label=f'Número ideal de clusters: {ideal_k}')
plt.legend()
plt.show()

# Justificativa
print(f"O número ideal de clusters foi identificado como {ideal_k}.")
print("Isso se deve ao fato de que, após esse ponto, a redução na inércia se torna menor, indicando que novos clusters não estão contribuindo significativamente para a redução da variabilidade dentro dos grupos. Portanto, três clusters representam uma boa divisão dos dados.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Simulando dados
np.random.seed(42)
X = np.random.rand(100, 2) * 100

# Definindo o número de clusters
n_clusters = ideal_k  # Número sugerido pelo método do cotovelo

# Aplicando KMeans
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Obter os rótulos dos clusters
labels = kmeans.labels_

# Visualizando os clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=300, marker='X', label='Centros dos Clusters')
plt.title(f'Clusterização KMeans com {n_clusters} Clusters', fontsize=16)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)
plt.legend()
plt.grid()
plt.show()

# Exibir os centros dos clusters
print("Centros dos Clusters:")
print(kmeans.cluster_centers_)
```

## **4. K-Means com Pré-processamento**

```python
from sklearn.preprocessing import StandardScaler

# Simulando dados
data = np.random.rand(200, 2) * 100

# Normalizando os dados
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Aplicando K-Means
kmeans_normalized = KMeans(n_clusters=4)
y_kmeans_normalized = kmeans_normalized.fit_predict(data_normalized)

# Visualizando os resultados
plt.figure(figsize=(8, 6))
plt.scatter(data_normalized[:, 0], data_normalized[:, 1], c=y_kmeans_normalized, cmap='plasma', s=100, alpha=0.6)
plt.scatter(kmeans_normalized.cluster_centers_[:, 0], kmeans_normalized.cluster_centers_[:, 1],
            s=300, c='red', marker='X', label='Centroides')
plt.title('K-Means Clustering com Dados Normalizados', fontsize=16)
plt.xlabel('Feature 1 (Normalizado)', fontsize=12)
plt.ylabel('Feature 2 (Normalizado)', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
